```python
import pandas as pd
import numpy as np
import random
import string
```

```python
def score():
    return np.random.randint(50, 100)
```

```python
def student():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
df = pd.DataFrame([[score() for i in range(4)] for i in range(10)], index=[student() for i in range(10)],columns=['국어','영어','수학','사회'])
```

```python
df.iloc[[0],:]
```

```python
df.loc[:,'국어':'영어']
```

```python
# 과학이 추가됨, 기존 학생들에게 각 랜덤으로추가
```

```python
df.loc[:,"과학"] = pd.Series([score() for i in range(10)], index=df.index)
df
```

```python
# 각학생들의 점수의 총합을 표현하는 컬럼 추가
```

```python
df.loc['총합계'] = df.sum()
df.loc[:,'합계'] = pd.Series([df.loc[df.index[i]].sum() for i in range(0, 11)], index=df.index)
df
```

```python
df.sort_values(by='합계', ascending=True)
```
